```python
def calculate_averages_in_range(zp0, zp1, dz, dh, wps, wpd, izb, izu):
    """
    Calculates the average of `wps` and `wpd` arrays within a specified height range.

    Args:
        zp0 (float): Lower bound of the height range.
        zp1 (float): Upper bound of the height range.
        dh (array-like): Height values corresponding to the data.
        wps (array-like): 3D array containing the first set of data.
        wpd (array-like): 3D array containing the second set of data.
        izb (int): Offset below reference height.
        izu (int): Offset above reference height.

    Returns:
        tuple: (wpsa, wpda), the calculated averages within the range.
    """

    it0 = find_tstep(zp0, dh)  # Assuming you have a function `find_tstep` defined
    it1 = find_tstep(zp1, dh)

    wpsa = np.zeros(izb + izu)
    wpda = np.zeros(izb + izu)

    for i in range(it0, it1):
        iz0 = int(dh[i, 1] / dz)  # Assuming 'dz' is defined elsewhere 
        wpsa += wps[iz0 - izb : iz0 + izu, ii, i]  # Assuming 'ii' is defined
        wpda += wpd[iz0 - izb : iz0 + izu, ii, i]

    wpsa /= it1 - it0
    wpda /= it1 - it0
    
    vt = (dh[it1, 1] - dh[it0, 1])/(dh[it1, 0] - dh[it0, 0])

    return wpsa, wpda, vt
```

```python
def calculate_energy_metrics(zp0, zp1, dz, dh, wlocal, dlam, izb, izu):
    """
    Calculates various metrics related to plastic deformation between positions zp0 and zp1.

    Args:
        zp0: Starting position along the z-axis.
        zp1: Ending position along the z-axis.
        dz: Spatial resolution along the z-axis.
        dh: 2D array containing spatial and temporal information (h-coordinates).
        wlocal: 2D array containing local width data at different z positions and time steps.
        dlam: Array used for index determination (assumed to be related to position).
        izb: Number of z-indices below the center to include in the wpa calculation.
        izu: Number of z-indices above the center to include in the wpa calculation.

    Returns:
        wmax: Averaged maximum local width in the specified region.
        zpl: Averaged plastic zone size in the specified region.
        wpa: Averaged projected area of the plastic zone.
        vt: Average pulling velocity in the specified region.
    """

    izp0, izp1 = find_indices_2d(dlam, 1e-12)  # Find indices corresponding to zp0 and zp1 in dlam
    it0, it1 = find_tstep(zp0, dh), find_tstep(zp1, dh)  # Find corresponding time steps

    # Extract relevant part of the local width data
    relevant_wlocal = wlocal[:, it0:it1]  

    # Calculate averaged maximum local width
    #wmax = np.mean(np.max(relevant_wlocal, axis=0))  

    # Calculate averaged plastic zone size
    zpl = dz * np.mean(izp1 - izp0) 

    # Calculate averaged projected area of the plastic zone (wpa)
    wpa = 0
    wmax = 0
    for i in range(it0, it1):
        iz0 = int(dh[i, 1] / dz)  # Calculate z-index at the center for time step i
        wpa += np.sum(wlocal[iz0 - izb:iz0 + izu, i]) * (dz * 1e3) ** 2  # Sum local widths in the vicinity and scale
        wmax += np.max(wlocal[iz0 - izb:iz0 + izu, i]) * (dz * 1e3) ** 2
        #wmax += np.max(wlocal[iz0 - izb:iz0 + izu, i]) / np.sum(wlocal[iz0 - izb:iz0 + izu, i])
    wpa /= (it1 - it0)  # Average over the time steps
    wmax /= (it1-it0)
    #wmax /= wpa

    # Calculate average pulling velocity
    vt = (dh[it1, 1] - dh[it0, 1]) / (dh[it1, 0] - dh[it0, 0])  # Change in h-position / change in time


    return wmax, zpl, wpa, vt
```

```python
# a function to calculate the non-dim 

def PoroFrac_scale(hs, phis, dp0):

    nu = 0.5 #0.2857
    G0 = 5e9
    gnu = G0 /(1-nu)
    rhog = 4900 * (4.9-dp0)/4.9 #DeltaRho*G
    ls = gnu/rhog
    
    zs = (phis*hs*ls)**0.5/1e3
    
    ks =  gnu * (phis * hs)**(3/4)/ls**(1/4) *2**(0.5)

    return zs, ks
```
